```python
rf_pipe = Pipeline([('preproc', prep),
                      ('rf', RandomForestClassifier(criterion= 'gini',
                                                    class_weight= 'balanced', random_state=123)  )
                  ])
```

**class_weight**

aparte de 'balanced', hay otras opciones que puedes utilizar para el parámetro class_weight en el clasificador SVM (Support Vector Machine) con kernel RBF ('rbf') en scikit-learn. Algunas de las opciones comunes incluyen:

None: No se aplica ponderación de clases. Todas las clases tienen el mismo peso.

'balanced' (equilibrado): Como ya mencionaste, este ajusta automáticamente los pesos de las clases inversamente proporcionales a sus frecuencias en los datos de entrenamiento.

dict: Puedes proporcionar un diccionario que asigna pesos específicos a cada clase. Por ejemplo, class_weight={0: 1, 1: 2} asignaría un peso de 1 a la clase 0 y un peso de 2 a la clase 1.

'balanced_subsample': Similar a 'balanced', pero utiliza un subconjunto de datos de entrenamiento para calcular los pesos de clase. Puede ser útil en conjuntos de datos grandes.

## Hiperparámetros

```python
rf_grid = dict(
    rf__ccp_alpha= [0.01, 0.03],
    rf__n_estimators= [50, 100],
    rf__max_features= [5, 8]  # En Python sí hacemos OHE
)
```

**rf__ccp_alpha**: Este hiperparámetro corresponde al parámetro de complejidad (Cost-Complexity Pruning, CCP) para el Random Forest. Es la penalización aplicada a la complejidad de cada árbol en el bosque. Un valor más alto de ccp_alpha generará árboles más pequeños y menos complejos, ayudando a prevenir el sobreajuste. En tu lista, estás probando con valores de 0.01 y 0.03.

**rf__n_estimators**: Este hiperparámetro especifica el número de árboles en el bosque. Un mayor número de árboles generalmente mejora la robustez y la capacidad de generalización del modelo, pero también puede aumentar el costo computacional. En tu lista, estás probando con 50 y 100 árboles.

**rf__max_features**: Este hiperparámetro controla la cantidad máxima de características que se deben considerar al dividir un nodo. Puede ser un número fijo o un porcentaje del total de características. Limitar el número de características puede ayudar a prevenir el sobreajuste
